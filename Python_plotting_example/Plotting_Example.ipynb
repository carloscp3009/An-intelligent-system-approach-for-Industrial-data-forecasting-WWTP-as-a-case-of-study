## Introduction to creating beautiful plots for your LaTeX thesis:

Python is a useful tool to create publication quality plots for your thesis. This example makes use of the package [SciencePlots](https://github.com/garrettj403/SciencePlots), which creates beautiful plots. 

It also sets up the figure sizes to match the column width of the latex template. This is important to maintain consistent font sizes. Further information on why this is important can be found, and the source of some of these settings can be found  [here](https://scipy-cookbook.readthedocs.io/items/Matplotlib_LaTeX_Examples.html).

Finally, '.pdf' is the recommended filetype for saving figures for your thesis. As they're essentially vector graphics they offer a much smaller filesize while maintaining high quality when compared to image files like .png or .jpg.

```python
import numpy as np    #This imports the NumPy package we will use for this example, to plot CSV files simply google plotting from csv python to find hundreds of examples, be use to include the settings below to get the correct figure sizes
import matplotlib.pyplot as plt  #This is the package we will use to plot



plt.style.use(['science','muted','no-latex']) #This choses the SciencePlots style, see the link above for other styles and colours

fig_width_pt = 438.17247   # If changing the margins from the template settings get the new colunmwidth from LaTeX using \showthe\columnwidth
                           # The column width will be returned the log output.
    
inches_per_pt = 1.0/72.27               # Convert pt to inches
golden_mean = (np.sqrt(5)-1.0)/2.0      # Aesthetic ratio
fig_width = fig_width_pt*inches_per_pt  # width in inches
fig_height =fig_width*golden_mean       # height in inches
fig_size = [fig_width,fig_height]
SMALL_SIZE = 4
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

plt.rc('font', size=10)          # controls default text sizes
plt.rc('axes', titlesize=10)     # fontsize of the axes title
plt.rc('axes', labelsize=10)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=8)    # fontsize of the tick labels
plt.rc('ytick', labelsize=8)    # fontsize of the tick labels
plt.rc('legend', fontsize=8)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.rcParams['text.usetex'] = True
plt.rcParams['figure.figsize'] = fig_size
plt.rcParams['figure.dpi'] = 200
plt.rcParams["errorbar.capsize"] = 3
```

## Example One

Single sine wave.

Note: For a thesis plot titles are not recommended, let your caption explain the plot.

```python
x = np.linspace(0,10, num=500)  #Creates an array from 0 to 10 with 500 evenly space data points
y = np.sin(x)                   # Takes the sine of the array we just created

plt.plot(x,y,label='Sine Wave') # Plots our x and y and labels it for our legend
plt.xlabel('x')                 # Label's x
plt.ylabel('Sin(x)')            # Label's y
plt.legend()                    # Creates legend based on labels
plt.savefig('example_1.pdf')    # Saves the plot
plt.show()                      # Shows the plot in the Jupyter notebook
```

## Example Two

Multiple Lines on a plot. In this case, sine and cosine waves.

Note: For a thesis plot titles are not recommended, let your caption explain the plot.

```python
x  = np.linspace(0,10, num=500)   #Creates an array from 0 to 10 with 500 evenly space data points as before
y  = np.sin(x)                    # Takes the sine of the array we just created as before
y2 = np.cos(x)                    # Takes the cosine of the array we created

plt.plot(x,y,label='Sine Wave')   # Plots our x and y and labels it for our legend
plt.plot(x,y2,label='Cosine Wave')# Plots our second line
plt.xlabel('x')                   # Label's x
plt.ylabel('Sin(x)/Cos(x)')       # Label's y
plt.legend()                      # Creates legend based on labels
plt.savefig('example_2.pdf')      # Saves the plot
plt.show()                        # Shows the plot in the Jupyter notebook
```

## Example Three

Scatter plots



```python
np.random.seed(243)
x = np.random.randint(10000, size=(50,2))   # Creates 2-D array of random values from 0-1


plt.scatter(x[:,0],x[:,1])        # Plots column 1 as x axis and coloumn 2 as y axis
plt.xlabel('x')                   # Label's x
plt.ylabel('y')                   # Label's y, matplotlib with accept some latex commands
plt.savefig('example_3.pdf')      # Saves the plot
plt.show()                        # Shows the plot in the Jupyter notebook
```

## Example Four

Using Tex in plots

```python
x = np.linspace(0,10, num=500)  #Creates an array from 0 to 10 with 500 evenly space data points
y = np.exp(x)                   # Takes the exponential of the array we just created

plt.plot(x,y) # Plots our x and y and labels it for our legend
plt.xlabel(r'$\mathcal{x}$')      # Label's x using tex math symbols
plt.ylabel(r'$e^{x}$')          # Label's y using tex math symbols
plt.savefig('example_4.pdf')    # Saves the plot
plt.show()                      # Shows the plot in the Jupyter notebook
```
